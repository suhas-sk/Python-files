```python
#To discard n.a. values
new_df=pd.read_excel("Book1.xlsx","Sheet1",names=["tickers","eps","revenue","price","people"],
                     na_values={"eps":['not available',"n.a."],
                    "revenue":["n.a.","not available",-1],
                    "people":[-1,"n.a.","not available"],
                                "price":[-1,"n.a.","not available"]
                    }  
                    )
new_df
```

```python
#using converters

def convert_people(cell):
    if cell=="n.a.":
        return "sam walton"
    else:
        return cell
    
def convert_eps(cell):
    if cell==" ":
        return "not available"
    return None

new_df = pd.read_excel("Book1.xlsx",converters={"people":convert_people})
new_df
```
