```python
from PIL import Image 
from collections import defaultdict

#Getting pixel info
def pixel_info(filepath):
    w,h=Image.open(filepath).size
    return w*h,w,h,
    
def color_info(filepath):
    img = Image.open(filepath)
    by_color = defaultdict(int)
    for pixel in img.getdata():
        by_color[pixel] += 1
    return by_color
  
def get_pixel_color(filename,x,y):
    image = Image.open(filename)
    pixel = image.load()
    return pixel[x,y]
    
print(f'{pixel_info("Figure_1.png")[0]} total pixel\n') #prints the total pixels in the image

print(f'{pixel_info("Figure_1.png")[1]} width\n') #prints width of the iamge in pixel

print(f'{pixel_info("Figure_1.png")[0]} height\n') #prints height of the image in pixels

print(color_info(r"E:\New folder\Figure_1.png"))  #prints dictionary color code for each pixel 
      
print(f'color of specified pixel is{ get_pixel_color("Figure_1.png",256,15)}') #Prints color of specified pixel
```
