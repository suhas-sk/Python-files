```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
flg = plt.figure()
ax = plt.axes(projection="3d")

x = [1,3,6,2,7,9,4,8,5] 
y = [9,6,3,5,2,7,1,4,8] 
z = [1,2,3,4,5,6,7,8,9]

ax.scatter(x,y,z,color="blue")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```
