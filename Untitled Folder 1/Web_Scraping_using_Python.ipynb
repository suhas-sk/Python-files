```python
import bs4
from urllib.request import urlopen as uReq
import requests
import smtplib
import pandas as pd

my_url = "https://www.flipkart.com/search?q=lenovo+laptops&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_6_na_na_na&as-pos=2&as-type=RECENT&suggestionId=lenovo+laptops%7CLaptops&requestId=1d51cfbd-0efb-49cf-bc69-8f627f79019c&as-backfill=on"

uclient = uReq(my_url)
page_html = uclient.read()                                 # extracting data from website
uclient.close()
page_soup = bs4.BeautifulSoup(page_html, "html.parser")    # parses data to code

containers = page_soup.findAll("div", {"class": "_3O0U0u"})

print("Total items=", len(containers))


container = containers[0]


file = open("File4.csv", "w")
headers="ProductName,Price,Discount\n"
file.write(headers)

for container in containers:                                 # loop to check and print product details

    product_name = container.div.img["alt"]
    print("Name:", product_name)

    price = container.findAll("div", {"class": "_1vC4OE _2rQ-NK"})
    price_container = price[0].text.strip()                    # strip() eliminates unnecessary commas,spaces,etc
    price_container = price_container.replace("â‚¹", "Rs.") 
    price_container = price_container.replace(",","")
    print("Price:", price_container)


    discount = container.findAll("div", {"class": "VGWI6T"})

    for discount_container in discount:
        discount_container = discount[0].text
        print("Discount:", discount_container)


    data = product_name + "," + price_container + ", " + discount_container + "\n"
    file.write(data)
    print("")

file.close()

    

    
```

```python
def send_mail():

    #NOTE: Turn off 'Less Secure App Access' of the google account from you are going to send the mail.
    
    
    
    # creates SMTP session
    s = smtplib.SMTP('smtp.gmail.com', 587)

    # start TLS for security
    s.starttls()
    
    # Authentication
    s.login("sender_email_id", "sender_email_id_password")

    # message to be sent
    message = "Price for the product fell down"
    # sending the mail
    s.send_mail("sender_email_id", "receiver_email_id", message)

    # terminating the session
    s.quit()
```
